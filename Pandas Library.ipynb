# correlations

### The corr() method calculates the <font color='red'>relationship</font>  between<font color='green'> each column </font>  in your data set.

#### The number varies from -1 to 1.

><font color='blue'>1</font> means that there is a **1 to 1 relationship (a perfect correlation)**, and for this data set, each time a value went up in the first column, the other one went up as well.

><font color='blue'>0.9</font> is also a **good relationship**, and if you increase one value, the other will probably increase as well.

><font color='blue'>-0.9</font> would be just as **good relationship** as 0.9, but if you increase one value, the other will probably go down.

><font color='blue'>0.2</font> means**NOT a good relationship**, meaning that if one value goes up does not mean that the other will.

```python
ds.corr()      # here i check the correlation of the data set that i created befor myslf.
```

```python
Stats.corr()           #here i check the correlation of the dataset that i created on finding stst functions collectively.
```
