```python
Data_1 = {                          # here we create our own dataset
    'A' : np.arange(40),
    'B' : np.random.randint(0,40,40),
    'C' : np.random.randn(40)
}
Data_1['D'] = Data_1['A']+5*np.random.randn(40)     # here we can add a new column like that in our dataset.

Data_1
```

# Plotting

```python
# for plot size (figure)
fig,ax = plt.subplots(figsize = (5,3.3), layout = 'constrained')     # here first we plot a figure only
```

```python
# for ploting data (Axes)
fig,ax = plt.subplots(figsize = (5,3.3), layout = 'constrained')
ax.scatter(Data_1['A'],Data_1['D'],Data_1['C'])    # method 1        
```

```python
fig,ax = plt.subplots(figsize = (5,3.3), layout = 'constrained')
ax.scatter(Data_1['A'],Data_1['D'],Data_1['C'])    # method 1
ax.scatter('A','D',data = Data_1)                  # method 2      # by using two methods combine
```

```python
fig,ax = plt.subplots(figsize = (5,3.2), layout = 'constrained')
ax.scatter(Data_1['A'],Data_1['D'],Data_1['C'])    # method 1
ax.scatter('A','D',data = Data_1)                  # method 2
ax.scatter(x = 'A',y = 'D',data = Data_1)          # method 3         # by using three methods combine
```

# Adding colors , size & Labels

```python
fig,ax = plt.subplots(figsize = (4,2.5), layout = 'constrained')    # first we must have to add the figure(background for plot)
ax.scatter(x= 'A',y = 'D',c = 'C', s = 'D', data = Data_1)      # in this c = color , s = size
# for adding labels
ax.set_xlabel('values of A')          # label for x axis
ax.set_ylabel('values of D')          # label for y axis
ax.set_title('Comparison of A and D values')      # label for title to show relation of both axis at the top.
```

```python
fig,ax = plt.subplots(figsize = (5,4.5), layout = 'constrained')    # first we must have to add the figure(background for plot)
ax.scatter(x= 'A',y = 'C', data = Data_1)                       # here i make another plot by changing its axis value
ax.set_xlabel('values of A')          # label for x axis 
ax.set_ylabel('values of C')          # label for y axis
ax.set_title('Comparison of A and C values')            # label for title     
```

```python
fig,ax = plt.subplots(figsize = (5,4.5), layout = 'constrained')    # first we must have to add the figure(background for plot)
ax.scatter(x= 'A',y = 'C',c = 'purple', s = 'B', data = Data_1)   # here i make another plot by changing its axis value and also add add color andsice.
ax.set_xlabel('values of A')
ax.set_ylabel('values of C')
ax.set_title('Comparison of A and C values')
```

# <font color='red'>Coding styles</font>
## <font color='greem'>Explicit style</font>

> in this style we can put **multiple data** at one axes.

> ***Explicity*** means that we have ***to state the type and value of a variable clearly and without ambiguity(not clearly understandable)***.

## <font color='orange'>Implicit style</font>

> ***Implicit*** means that ***Python automatically infers the type and value of a variable without stating it explicitly***.

```python
# Explicit style
# with ONE data
x = np.linspace(0,3,300)
fig,ax = plt.subplots(figsize = (5,3.4), layout = 'constrained')    # here first we adjust a figure( background) by fixing its figsize ourself.
ax.plot(x , x , label = 'linear')        # here it will plot a graph with a straight(linear) line in it. also a first data.(blue line)
ax.plot(x , x**2 , label = '2nd line')    # here it # here it will add a slightly curved line in it. also it's a second data.( orange line)
ax.plot(x , x**3 , label = '3rd line')     # here it # here it will add a new curved line in it. also it's a third data.(green line)
ax.plot(x , x**4 , label = '4th line')      # here it # here it will add a another curved line in it. also it's a forth data.( red line)
ax.set_xlabel('x values')                    # Label for x-axis.
ax.set_ylabel('x and their power(square)')    # Label foe y-axis.
ax.set_title('Line plot')                      # Title ( to give information about the plot).
ax.legend()                                     # to add a square 'markout' to give information about the data that plot show as a graph.
```

```python
# Another example ( with TWO data)
x = np.linspace(0,3,300)
y= np.linspace(0,1,300)
fig,ax = plt.subplots(figsize = (5,3.4), layout = 'constrained')    # here first we adjust a figure( background) by fixing its figsize ourself.
ax.plot(x , y , label = 'linear')            # here it will plot a graph with a straight(linear) line in it. also a first data.(blue line)
ax.plot(x**2 , y**2 , label = '2nd line')     # here it will add a line in it. also it's a second data.(orange line)
ax.plot(x**3 , y**3 , label = '3rd line')      # here it will add a new  line in it. also it's a third data.(green line)
ax.plot(x**4 , y**4 , label = '4th line')       # here it will add a another  line in it. also it's a forth data.(red line)
ax.set_xlabel('X values')                        # Label for x-axis.
ax.set_ylabel('Y values')                       # Label for y-axis.
ax.set_title('Line plot')                      # Title ( to give information about the plot).
ax.legend()                                   # to add a square 'markout' to give information about the data that plot show as a graph.
```

```python
# implicit style
# with ONE data
x = np.linspace(0,3,300)
plt.subplots(figsize = (5,3.4), layout = 'constrained')    # here first we adjust a figure( background) by fixing its figsize ourself.
plt.plot(x , x , label = 'linear')        # here it will plot a graph with a straight(linear) line in it. also a first data.(blue line)
plt.plot(x , x**2 , label = '2nd line')    # here it # here it will add a slightly curved line in it. also it's a second data.( orange line)
plt.plot(x , x**3 , label = '3rd line')     # here it # here it will add a new curved line in it. also it's a third data.(green line)
plt.plot(x , x**4 , label = '4th line')      # here it # here it will add a another curved line in it. also it's a forth data.( red line)
plt.xlabel('x values')                        # Label for x-axis.
plt.ylabel('x and their power(square)')        # Label foe y-axis.
plt.title('Line plot')                          # Title ( to give information about the plot).
plt.xlim(0,3)                                    # limitation for x value
plt.legend()                                     # to add a square 'markout' to give information about the data that plot show as a graph.
```

```python
# Another example ( with TWO data)
x = np.linspace(0,3,300)
y= np.linspace(0,1,300)
plt.subplots(figsize = (5,3.4), layout = 'constrained')    # here first we adjust a figure( background) by fixing its figsize ourself.
plt.plot(x , y , label = 'linear')            # here it will plot a graph with a straight(linear) line in it. also a first data.(blue line)
plt.plot(x**2 , y**2 , label = '2nd line')     # here it will add a line in it. also it's a second data.(orange line)
plt.plot(x**3 , y**3 , label = '3rd line')      # here it will add a new  line in it. also it's a third data.(green line)
plt.plot(x**4 , y**4 , label = '4th line')       # here it will add a another  line in it. also it's a forth data.(red line)
plt.xlabel('X values')                            # Label for x-axis.
plt.ylabel('Y values')                             # Label for y-axis.
plt.title('Line plot')                            # Title ( to give information about the plot).
plt.xlim(0,3)                                    # limitations for x values
plt.ylim(0,10)                                  # limitations for y values
plt.legend()                                   # to add a square 'markout' to give information about the data that plot show as a graph.
```

# 1. <font color='purple'>Bar plot</font>
##   The bar plots are <font color='coral'>vertical/horizontal rectangular graphs</font> that show data comparison.

> you can gauge(arrange) the changes over a period represented in another axis (mostly the X-axis).

> Each bar can store the value of one or multiple data divided in a ratio.

> The longer a bar becomes, the greater the value it holds.

```python
plt.bar([1,2,3,4],[30,40,20,60],label="Carpenter",color='coral',width=0.5)
plt.bar([0.75,1.75,2.75,3.75],[40,50,30,70],label="Plumber", color='purple',width=0.5)
plt.xlabel('Days')
plt.ylabel('Working hours')
plt.title('Bar plot')
plt.legend()

```

# <font color='grey'>Scatter plot</font>
###      scatter plots is used to compare various data variables to <font color='pink'>determine the connection </font> between dependent and independent variables.

> The data gets expressed as a collection of points clustered together meaningfully. 

>  Here each value has one variable (x) determining the relationship with the other (Y).

```python
x1 = [1,4,3,2,5]
y1 = [1,3,1,7,10]
x2=[6,7,8,9,10]
y2=[3.5,8.5,6.5,4.5,10.5]
plt.scatter(x1, y1, label = 'high bp low heartrate', color='grey')
plt.scatter(x2,y2,label='low bp high heartrate',color='pink')
plt.title('Smart band Report')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
```

# <font color='teal'>Histogram plot</font> 
###   A histogram is an approximate representation of the <font color='red'>distribution of numerical data</font>.

> Histograms may have a **similar appearance to bar charts** but unlike them, are used to plot the frequency of variable occurrences in continuous data.

> This continuous data is divided into classes called **bins**. The **majority and minority** points can be easily spotted using these bins.

> Both histogram and bar plot look alike but are used in **different scenarios**. 

```python
pop = [22,55,62,45,21,22,34,42,42,4,2,8]
plt.hist(pop, bins = 5, rwidth=0.4 , color = 'teal')                  
plt.xlabel('Age groups')
plt.ylabel('Number of people')                                   
plt.title('Histogram plot')
```

# <font color='green'>Line plot</font>
### Line plots are drawn by joining <font color='fuschia'>straight lines connecting data points</font> where the x-axis and y-axis values intersect.

> Line plots are the simplest form of representing data.

```python
x = np.array([1,2,3,5,6])
y = [0, 7, 0,4,0]
plt.plot(x, y , color = 'green')
```

# <font color='orange'>Box plot</font>
###   Box Plot is also known as <font color='green'>Whisker plot</font> is created to display the summary of the set of data.

> Data values having properties like **minimum, first quartile, median, third quartile and maximum**.

> In the box plot, a box is created from the **first quartile to the third quartile**.

> A **vertical line** is also there which goes **through the box at the median**. 

> **x-axis** denotes the **data to be plotted** while the **y-axis** shows the **frequency distribution.**

```python
data = np.random.normal(100, 20, 200)
plt.figure(figsize =(10, 7))
plt.boxplot(data)
plt.show()
```
