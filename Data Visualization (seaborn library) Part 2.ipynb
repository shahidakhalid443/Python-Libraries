## by default  scatter plots will form

```python
sns.relplot(data = puffins , x = "body_mass_g" , y = "species")
plt.show(0)     # here a scatter plot is formed by default this plot is descibing the body mass of penguins on the basis of different species of penguins.
```

# <font color='teal'>Scatter plot</font>
## A scatter plot (aka scatter chart, scatter graph) uses <font color='red'>dots to represent values</font> for two different numeric variables.

> The position of each dot on the **horizontal and vertical** axis indicates values for an individual data point.

> Scatter plots are used to **observe relationships between variables**.

> A scatter plot can also be useful for **identifying other patterns** in data. We can divide data points into groups based on   how closely sets of points cluster together.

> Scatter plots can also show if there are any **unexpected gaps** in the data and if there are any **outlier points**. 

```python
sns.scatterplot(data = puffins , x = "bill_depth_mm" , y = "bill_length_mm")      # here we make a simple scatter plot betweenbil_length and bill_width of the penguins.
plt.xlabel("Bill_Depth of puffins")              # here we add a label Boll_Depth of puffins at X-axis 
plt.ylabel("Bill_Length of puffins")               # here we add a label Bill_length of puffins at y-axis.
plt.title("Bill Depth & Length of puffins")              # here we add a title Bill depth & length of puffins at the top of the plot just to make it more attractive.
plt.show(0)
```

```python
sns.set_style("whitegrid")         # here we add set_style function which can change the background grid style.
sns.scatterplot(data = puffins , x = "bill_depth_mm" , y = "bill_length_mm" , hue = "species" )   # here we make a simple scatter plot betweenbil_length and bill_width of the penguins.
            # also we add ( hue = )which will color to the plot on the basis of kind of species.
plt.xlabel("Bill_Depth of puffins")            # here we add a label Boll_Depth of puffins at X-axis 
plt.ylabel("Bill_Length of puffins")            # here we add a label Bill_length of puffins at y-axis.
plt.title("Bill & Length of puffins")            # here we add a title Bill depth & length of puffins at the top of the plot just to make it more attractive.
plt.show()
```

```python
sns.scatterplot(data = puffins , x = "bill_depth_mm" , y = "bill_length_mm" , hue = "species", style = "species" )# here we make a simple scatter plot betweenbil_length and bill_width of the penguins.
            # also we add ( hue = )which will color to the plot on the basis of kind of species.
             # also we add (style = ) which will show the dots onn scatter plot of different styles on the basis of the species types.
plt.xlabel("Bill_Depth of puffins")             # here we add a label Boll_Depth of puffins at X-axis
plt.ylabel("Bill_Length of puffins")             # here we add a label Bill_length of puffins at y-axis.
plt.title("Bill & Length of puffins")             # here we add a title Bill depth & length of puffins at the top of the plot just to make it more attractive.
plt.show()
```

```python
sns.scatterplot(data = puffins , x = "bill_depth_mm" , y = "bill_length_mm" , hue = "species", style = "species" , size = "sex" )
           # here we make a simple scatter plot betweenbil_length and bill_depth of the puffins (penguins).
            # also we add ( hue = )which will color to the plot on the basis of kind of species.
             # also we add (style = ) which will show the dots onn scatter plot of different styles on the basis of the species types.
              # here we also add (sixe = ) that will repesent no of the puffins(penguins) types are present on the basis of sex(Male , Female).
plt.xlabel("Bill_Depth of puffins")          # here we add a label Boll_Depth of puffins at X-axis.   
plt.ylabel("Bill_Length of puffins")           # here we add a label Bill_length of puffins at y-axis.
plt.title("Bill & Length of puffins")           # here we add a title Bill depth & length of puffins at the top of the plot just to make it more attractive.
plt.show()
```

# <font color='purple'>Line plot</font>
###   a graph that displays data as points or check marks above a number line, showing the frequency of each value. 

> Many **parameters are used to control what visual semantics are used to identify the different subsets**.

> It is possible to show up to **three dimensions independently** by using all three semantic types, but this style of plot can be hard to interpret and is often ineffective. 

```python
sns.lineplot(data = puffins , x = "bill_depth_mm" , y = "bill_length_mm")
             # here we plot a simple line plot on the basis of bill_depth and bill_length.
```

```python
sns.lineplot(data = puffins , x = "bill_depth_mm" , y = "bill_length_mm" , hue = "species" ,
             style = "sex" , size = "sex")
      # here we make a simple line plot between bill_depth and bill_length of the puffins (penguins).
       # also we add ( hue = )which will color to the plot on the basis of kind of species.
        # also we add (style = ) which will show the dots on line plot of different styles on the basis of the sex(male , female)types.
         # here (size = ) gives no of poffins on the basis of sex(male , female) be bolding the lines on line plot.
            
plt.xlabel("bill depth of puffins")   # here we add a x label just to make our graph attractive.
plt.ylabel("bill length of puffins")    # here we add a y label just to make our graph attractive.
plt.title("line plot")                   # here we add a title on the top just to make our graph attractive and easy to recognizeable..
```

```python
sns.lineplot(data = puffins , x = "bill_depth_mm" , y = "bill_length_mm" , hue = "species" ,
             style = "sex" , size = "sex", err_style="bars")
      # here we make a simple line plot between bill_depth and bill_length of the puffins (penguins).
       # also we add ( hue = )which will color to the plot on the basis of kind of species.
        # also we add (style = ) which will show the dots on line plot of different styles on the basis of the sex(male , female)types.
         # here (size = ) gives no of poffins on the basis of sex(male , female) be bolding the lines on line plot.
           # here (err_style = )add bars that draw the confidence intervals with translucent error bands or discrete error bars. 
plt.xlabel("bill depth of puffins")   # here we add a x label just to make our graph attractive.
plt.ylabel("bill length of puffins")    # here we add a y label just to make our graph attractive.
plt.title("line plot")                   # here we add a title on the top just to make our graph attractive and easy to recognizeable.
```

```python
sns.lineplot(data = puffins , x = "bill_depth_mm" , y = "bill_length_mm" , hue = "species" ,
             style = "sex" , size = "sex", err_style="bars", estimator=None , legend = "brief")
      # here we make a simple line plot between bill_depth and bill_length of the puffins (penguins).
       # also we add ( hue = )which will color to the plot on the basis of kind of species.
        # also we add (style = ) which will show the dots on line plot of different styles on the basis of the sex(male , female)types.
         # here (size = ) gives no of poffins on the basis of sex(male , female) be bolding the lines on line plot.
           # here (err_style = )add bars that draw the confidence intervals with translucent error bands or discrete error bars. 
            # here (estimator = True) Method for aggregating across multiple observations of the y variable at the same x level.If None, all observations will be drawn.
             # legend If brief, numeric hue and size variables will be represented with a sample of evenly spaced values. 
plt.xlabel("bill depth of puffins")   # here we add a x label just to make our graph attractive.
plt.ylabel("bill length of puffins")    # here we add a y label just to make our graph attractive.
plt.title("line plot")                   # here we add a title on the top just to make our graph attractive and easy to recognizeable.
```

# 2.  displots (Distributional)  
## Making plot using category (tag)


```python
sns.set_style("darkgrid")    # here we use a backgroun theme to make graphs more attractive.
sns.displot(data = puffins , x = "bill_length_mm" , hue = "species" , kde = True )   # here we plot a dist plot that is by default a histogram plot on the basis of bill length of puffins(penguins)
           # here (hue =  ) gives color to the graph on the basis of species,
            # here ( kde = True) shows distribution in the Data.
plt.title("Histogram plot")      # here we gives a title to the plot to make it attractive.
plt.show()
```

```python
sns.set_style("darkgrid")
sns.displot(data = puffins , x = "bill_length_mm" , kde = "species" )   # Here its the same plot with out colors the only difference is (kde = ) is giving distribution on the basis of species.
plt.title("Histogram plot")            # here we gives a title to the plot to make it attractive.
plt.show()
```

```python
sns.set_style("darkgrid")
sns.displot(data = puffins , x = "bill_length_mm" , hue = "species" , kind = "kde" , col = "sex")  # here we plt a graph on the basis of bill length of puffins(penguins)
                       # here (hue = ) function gives color on the basis of species.
                        # here (kind = ) in this function we gives kind of displot to plot that type of plot which we want.
                         # here (col = ) function will make two graph on the basis of sex( male , female)
plt.title("kde plot")     # here we gives title to our plot.
plt.show(0)
```

```python
sns.set_style("darkgrid")
sns.displot(data = puffins , x = "bill_length_mm" , hue = "species" , kind = "ecdf")      # here we plt a graph on the basis of bill length of puffins(penguins)
                       # here (hue = ) function gives color on the basis of species.
                        # here (kind = ) in this function we gives kind of displot to plot that type of plot which we want.
plt.show(0)
```
