# <font color='teal'>Scatter plot</font>
## A scatter plot (aka scatter chart, scatter graph) uses <font color='red'>dots to represent values</font> for two different numeric variables.

> The position of each dot on the **horizontal and vertical** axis indicates values for an individual data point.

> Scatter plots are used to **observe relationships between variables**.

> A scatter plot can also be useful for **identifying other patterns** in data. We can divide data points into groups based on   how closely sets of points cluster together.

> Scatter plots can also show if there are any **unexpected gaps** in the data and if there are any **outlier points**. 

```python
sns.scatterplot(data = puffins , x = "bill_depth_mm" , y = "bill_length_mm")      # here we make a simple scatter plot betweenbil_length and bill_width of the penguins.
plt.xlabel("Bill_Depth of puffins")              # here we add a label Boll_Depth of puffins at X-axis 
plt.ylabel("Bill_Length of puffins")               # here we add a label Bill_length of puffins at y-axis.
plt.title("Bill Depth & Length of puffins")              # here we add a title Bill depth & length of puffins at the top of the plot just to make it more attractive.
plt.show(0)
```

# Histogram
## A histogram is a classic visualization tool that represents the distribution of one or more variables by counting the number of observations that fall within discrete bins.

> This function can normalize the statistic computed within each bin to estimate frequency,  probability mass, and it can add a smooth curve obtained using a kernel density estimate

```python
sns.histplot(data = puffins , x = "body_mass_g", bins = 20)     # here we plot a simpe histogram on the basis of body mass of puffins(penguins)
                          # here (bin = )we can specify no of bins by ourself. 
plt.show()
```

```python
sns.histplot(data = puffins , x = "body_mass_g", bins = 20 , fill = False )     # here we plot a simpe histogram on the basis of body mass of puffins(penguins)
                          # here (bin = )we can specify no of bins by ourself. 
                           # here ( fill = False) you can show unfilled bars
plt.xlabel("Body masss of puffins")         # here we can also add labels to make the plot more attractive.
plt.title("Histogram plot")                  # here we can also add title at the top to make it more easy to recogize.
plt.show()
```

# ecdf plot (Empirical Commutative Distribution)
###  ECDF plot visualizes each and every data point of the dataset directly, which makes it easy for the user to interact with the plot.

> This plot contains more information because it has no bin size setting, which means it doesn’t have any smoothing parameters.

> Since its curves are monotonically increasing, so it is well suited for comparing multiple distributions at the same time.

> The x-axis corresponds to the range of values for the variable 

> The y-axis corresponds to the proportion of data points that are less than or equal to the corresponding value of the x-axis.

# rug plot 

###  kind must be one of ['hist', 'kde', 'ecdf'], but rug was passed.

> used to plot **marginal distributions** and allows to draw ticks along the x and y axes.

> A rug plot is a graph that plots data for **single quantitative values** that is displayed as marks along the axes.

> These plots are usually used for **two-dimensional scatter plots** by marking the rug plot of x values along the x-axis and rug plot of y values along the y-axis.

```python
sns.set_style("darkgrid")
sns.rugplot(data = puffins , x = "bill_length_mm" , y = "bill_depth_mm" , hue = "species" , height = .08 )
plt.title("Rug plot")      # here we can also add title at the top to make it more easy to recogize.
```
